```
item_features.groupBy('item_id').agg(
    count('item_id').alias('unique_items')
).count()
```

```
item_per_category = item_features.groupBy(
    'feature_category_id'
).agg(
    count('item_id').alias('item_per_category')
).orderBy(
    col('item_per_category').desc()
)

item_per_category.display()
```
